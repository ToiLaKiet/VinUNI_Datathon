```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv("../../sales.csv")
```

```python
df
```

```python
df_rev_prophet = df[['Date', 'Revenue']].rename(columns={'Date': 'ds', 'Revenue': 'y'})
```

```python
df_rev_prophet
```

```python
m = Prophet(
    growth='linear',                  # or 'logistic' if revenue has a ceiling
    yearly_seasonality=True,
    weekly_seasonality=True,
    seasonality_mode='multiplicative', # crucial for revenue — variance grows with level
    changepoint_prior_scale=0.05      # controls trend flexibility, tune this
)
```

```python
m.fit(df_rev_prophet)
```

```python
future = pd.read_csv('../../sample_submission.csv')[['Date']].rename(columns={'Date': 'ds'})
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
ans = pd.read_csv('../../sample_submission.csv')
```

```python
ans['Revenue'] = forecast['yhat'].values
```
