```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('../../orders.csv')
order_items = pd.read_csv('../../order_items.csv')
products = pd.read_csv('../../products.csv')
customers = pd.read_csv('../../customers.csv')
```

```python
orders_detail = orders.merge(order_items, on='order_id', how='left').merge(products, on='product_id', how='left').drop(columns=['price'])
orders_detail
```

```python
orders_detail.columns
```

```python
orders_detail['total_revenue'] = orders_detail['quantity'] * orders_detail['unit_price']
```

```python
orders_detail
```

```python
orders_detail['total_cogs'] = orders_detail['quantity'] * orders_detail['cogs']
orders_detail
```

```python
orders_detail.drop(columns=['quantity', 'unit_price','cogs'], inplace=True)
```

```python
orders_detail['order_date'] = pd.to_datetime(orders_detail['order_date'])   
```

```python
orders_detail
```

```python
promotions = pd.read_csv('../../promotions.csv')
orders_detail = orders_detail.merge(
    promotions[['promo_id', 'discount_value']].rename(columns={'discount_value': 'dis_1'}),
    on='promo_id',
    how='left'
)
orders_detail.drop(columns=['promo_id'], inplace=True)
orders_detail
```

```python
orders_detail = orders_detail.merge(
    promotions[['promo_id', 'discount_value']].rename(columns={'discount_value': 'dis_2'}),
    left_on='promo_id_2',
    right_on='promo_id',
    how='left'
)
orders_detail.drop(columns=['promo_id_2','promo_id'], inplace=True)
orders_detail
```

```python
orders_detail = orders_detail.merge(customers[['customer_id', 'city']], on='customer_id', how='left').drop(columns=['customer_id'])
orders_detail
```

```python
orders_detail['dis_1'] = orders_detail['dis_1'].fillna(0)
orders_detail['dis_2'] = orders_detail['dis_2'].fillna(0)
```

```python
df = orders_detail.groupby(['order_date','segment','category','order_source','dis_1','dis_2', 'city'])['total_revenue'].sum().reset_index()
df['total_cogs'] = orders_detail.groupby(['order_date','segment','category','order_source','dis_1','dis_2', 'city'])['total_cogs'].sum().values
```
