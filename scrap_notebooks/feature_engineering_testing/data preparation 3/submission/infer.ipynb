```python
import pandas as pd
```

```python
cogs_test = pd.read_csv('../cogs_test_prophet_seasonality.csv')
rev_test = pd.read_csv('../rev_test_prophet_seasonality.csv')
```

```python
cogs_test.drop('Date', axis=1, inplace=True)
rev_test.drop('Date', axis=1, inplace=True)
```

```python
import joblib
#Load
cogs_model = joblib.load('../cogs_model_prophet_seasonality.pkl')
rev_model = joblib.load('../rev_model_prophet_seasonality.pkl')
```

```python
ans = pd.read_csv('../cogs_test_prophet_seasonality.csv')[['Date']].copy()
ans['COGS'] = cogs_model.predict(cogs_test)
ans['Revenue'] = rev_model.predict(rev_test)
ans.to_csv('submission.csv', index=False)
```
