```python
import pandas as pd
import numpy as np
```

```python
orders = pd.read_csv('../../orders.csv')
order_items = pd.read_csv('../../order_items.csv')
products = pd.read_csv('../../products.csv')
customers = pd.read_csv('../../customers.csv')
geography = pd.read_csv('../../geography.csv')
```

```python
orders_detail = orders.merge(geography, on='zip', how='left').merge(order_items, on='order_id', how='left').merge(products, on='product_id', how='left')
orders_detail
```

```python
orders_detail['order_date'] = pd.to_datetime(orders_detail['order_date'])
```

```python
orders_detail['total_revenue'] = orders_detail['quantity'] * orders_detail['unit_price']
orders_detail['total_cogs'] = orders_detail['quantity'] * orders_detail['cogs']
orders_detail
```

```python
orders_detail.drop(columns=['quantity', 'unit_price','cogs'], inplace=True)
```

```python
orders_detail
```

```python
df = orders_detail.groupby(['order_date','city'])['total_revenue'].sum().reset_index()
df['total_cogs'] = orders_detail.groupby(['order_date','city'])['total_cogs'].sum().values
df
```

```python
df[df['city'] == 'Ho Chi Minh City']
```

```python
# lấy danh sách city
cities = df['city'].unique()

# loop từng city
for city in cities:
    temp = df[df['city'] == city].copy()
    
    # set index
    temp = temp.set_index('order_date').asfreq('D').fillna(0)
    
    # lưu file
    temp.to_csv(f'daily_{city}.csv')
```
