```python
import pandas as pd
import numpy as np
from prophet import Prophet
```

# Revenue Prophet

```python
import os

files = [f for f in os.listdir() if f.startswith('daily_')]

for file in files:
    df = pd.read_csv(file)
    
    df['order_date'] = pd.to_datetime(df['order_date'])
    
    df.drop('total_cogs', axis=1, inplace=True)
    
    df_prophet = df.rename(columns={
        'order_date': 'ds',
        'total_revenue': 'y'
    })
    
    model = Prophet(yearly_seasonality=True, weekly_seasonality=True)
    model.fit(df_prophet)
    
    future = model.make_future_dataframe(periods=548, freq='D')
    forecast = model.predict(future)
    
    features = forecast[['ds', 'trend', 'yearly', 'weekly']]
    
    df_final = df_prophet.merge(features, on='ds', how='right')
    
    city = file.replace('daily_', '').replace('.csv', '')
    df_final.to_csv(f'revenue_prophet/{city}_prophet_features.csv', index=False)
```
